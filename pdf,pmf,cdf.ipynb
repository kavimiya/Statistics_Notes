## Random Variable :

`A random  variable is the result of a chance event, that you may measure or count`. `It is a variable that assumes numerical values associated with the random outcomes of an experiment where one (and only one) numerical value is assigned to each sample point`. Let us consider the example of flipping a coin once, wherein the possible outcome is either a head or tail denoted by 1 and 0 respectively. Another example is the random possible outcome while rolling a dice which is one among the following (1,2,3,4,5,6).The random variable is of two types viz.,

* Discrete Random Variable
* Continuous Random Variable

## Discrete Random Variable :
A discrete random variable is variable which takes the numerical outcome of the chance event or any countable number of possible values. An example is the number of people visiting the hospital in a day. It can never be a decimal value.

## Continuous Random Variable :
A random variable is said to be a continuous random variable if it takes infinite number of values in an interval. The probability of the random variable taking any one specific value is zero.The random variable "X" can take any possible values in an interval of real numbers. For example X âˆˆ [0,1].

## What is Probability Distribution & Why?
Probability Distribution is a function which lists all the possible outcomes a random variable can take, along with its corresponding probability of occurrence. Although the random variable has the ability to assume a random value it lacks the capability to capture the probability of occurrence those value. The probability distribution can also be plotted as a graph.

## Probability Density Function :
The probability density function is the statistical function that defines the probability distribution of a `continuous random variable`. When the probability density function is graphically plotted, the area under the curve will indicate the interval in which the variable will fall. The total area in this interval of the graph equals the probability of a discrete random variable occurring. The sum of all probabilities should be equal to 1. `It is impossible to find the probability of a particular continuous random variable since the summation of probability of all random variables within an interval will be infinite.`

## PDF (Probability Density Function):

* PDF is used for continuous random variables.
* It gives the relative likelihood of the random variable taking on a particular value within a given range.
* Unlike PMF, PDF doesn't directly give probabilities but rather densities. The probability of a continuous random variable falling within a specific range is given by the integral of the PDF over that range.
* For example, if you have a normally distributed continuous variable (like height or weight), the PDF would describe the likelihood of observing different values within a given range.
